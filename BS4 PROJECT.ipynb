```python
#importing BeautifulSoup
from bs4 import BeautifulSoup
soup= BeautifulSoup(page.content,'html.parser')
```

```python
print(soup)
#this prints the soup
```

```python
#prettify function makes it easier to read,makes tag easier to see
print(soup.prettify())
```

```python
#moving through the structure one level at a time,using children property of soup
# a child is a tag inside another tag
list(soup.children)
#this tells us there are two tags at the top level of the page-<!doctype html> tag and the <html> tag
```

```python
#seeing what type of each element in the list are
[type(item) for item in list(soup.children)]
#from the response all items are beautifulsoup objects
# doctype object(information about the type of document),NavigableSting object(represents text found the in html document),tag object(contains other nested tags)
```

```python
#we can now select the html tag and its children by taking the third term in the list
html= list(soup.children)[2]
```

```python
#finding the children inside the html tag
list(soup.children)
#there are two tags here i.e head and body.lets extract the text inside the p tag so we can dive into the body
```

```python
body=list(soup.children)[3]
#now we can get the p tag by finding the children of the body tag
```

```python
list(body.children)
```

```python
#now we can isolate the p tag
p = list(body.children)[1]
```

```python
#after isolating the p tag we can use the get_text method to extract all the text inside the tag
p.get_text()
```

```python
#finding all instances of a tag at once the codes w ran above were useful for figuring out how to navigate the page
#if we want to extract a single tag,we can use the find_all method,this finds all the tags on a page
soup= BeautifulSoup(page.content,'html.parser')
soup.find_all('p')
#find_all function returns a list so we loop through it to extract text
```

```python
#looping through
soup.find_all('p')[0].get_text()
```

```python
#if you want to find the first instance of a tag you can use the find method which will return a bs object
soup.find('p')
```
