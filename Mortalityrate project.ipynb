```python
# finding the missing values
health_df['VALUE'].fillna(health_df['VALUE'].mean(), inplace=True)
health_df
```

```python
#dropping empty tables 
health_df.drop('STATUS',axis=1)
```

```python
health_df.drop('SYMBOL',axis=1)
```

```python
# Aggregate data
grouped = health_df.groupby(['Sex', 'Cause of death (ICD-10)'])['VALUE'].sum().reset_index()
top_causes_male = grouped[grouped['Sex'] == 'Males'].nlargest(5, 'VALUE')
top_causes_female = grouped[grouped['Sex'] == 'Females'].nlargest(5, 'VALUE')
```

```python
# Combine for grouped bar chart
top_combined = pd.concat([top_causes_male, top_causes_female])
```

```python
# Plot
sns.barplot(data=top_combined, x='Cause of death (ICD-10)', y='VALUE', hue='Sex')
plt.title('Top 5 Causes of Death by Gender')
plt.xticks(rotation=90)
plt.show()
```
